```python
from datetime import datetime
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 60
import re, os
```

```python
adds=pd.ExcelFile('Y:/CRT CEP Evaluation/Documents/CRP Eligible Buildings May 2017 FINAL.xls - CRP Eligible.xlsx')
print(adds.sheet_names)
df = adds.parse('CRP Eligible Buildings')
df.head()
```

```python
pat= r"([0-9\-/]+)(.*)"
repl1 = lambda m: m.group(1)
repl2 = lambda m: m.group(2)
df['number']=df['Edited_ADDRESS'].str.replace(pat, repl1)
pat2= r"([0-9]+)(\-[0-9/]+)?"
df['number']=df['number'].str.replace(pat2, repl1)
df['street']=df['Edited_ADDRESS'].str.replace(pat, repl2)
df['boro']=1

df['street'][df['street']==' BATTERY PLACE N']=' BATTERY PLACE'

df[['number','street','Edited_ADDRESS']].head(200)
```

```python
writer = pd.ExcelWriter('Y:/CRT CEP Evaluation/Documents/CRP Eligible Buildings May 2017 FINAL_Pythoned.xlsx')
df.to_excel(writer,'Pythoned')
writer.save()
```
